```python
from scipy.interpolate import lagrange
import numpy as np

# Define the points for which you want to find the Lagrange polynomial
x = np.array([1, 2, 3, 6])
y = np.array([1, 4, 9, 6])

# Use the lagrange function to find the polynomial
poly = lagrange(x, y)

print(poly)
print(poly.coefficients)
```
