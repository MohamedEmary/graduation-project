```python
import pandas as pd
import matplotlib.pyplot as plt
import scienceplots

plt.rcParams.update({'font.size': 14})
plt.style.use(['science', 'no-latex'])
plt.rcParams['font.family'] = 'Times New Roman'

bisection = pd.read_csv('../results/bisection-table.csv')
false_position = pd.read_csv('../results/false-position-table.csv')
hybridBF = pd.read_csv('../results/hybridBF-table.csv')
hybridSF = pd.read_csv('../results/hybridSF-table.csv')
secant = pd.read_csv('../results/secant-table.csv')
```

## Bisection, False Position, HybridBF


```python
plt.figure(figsize=(18, 6))
plt.plot(bisection['Problem'], bisection['Iter'], label='Bisection')
plt.plot(false_position['Problem'],
         false_position['Iter'], label='False Position')
plt.plot(hybridBF['Problem'], hybridBF['Iter'], label='HybridBF')
plt.legend()
plt.xlabel('Problem')
plt.ylabel('Iterations')
plt.title('Iteration Comparison')
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.5)
plt.tick_params(axis='x', which='minor', bottom=False)
plt.savefig('iterations-comparison-BF.svg')
plt.show()
```

```python
plt.figure(figsize=(18, 6))
plt.plot(bisection['Problem'], bisection['Avg CPU Time'], label='Bisection')
plt.plot(false_position['Problem'],
         false_position['Avg CPU Time'], label='False Position')
plt.plot(hybridBF['Problem'], hybridBF['Avg CPU Time'], label='HybridBF')
plt.legend()
plt.xlabel('Problem')
plt.ylabel('Time (s)')
plt.title('CPU Time Comparison')
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.5)
plt.tick_params(axis='x', which='minor', bottom=False)
plt.savefig('CPU-time-comparison-BF.svg')
plt.show()
```

```python
plt.figure(figsize=(18, 6))
plt.plot(bisection['Problem'], bisection['Function Value'], label='Bisection')
plt.plot(false_position['Problem'],
         false_position['Function Value'], label='False Position')
plt.plot(hybridBF['Problem'], hybridBF['Function Value'], label='HybridBF')
plt.legend()
plt.xlabel('Problem')
plt.ylabel('Function Value')
plt.title('Function Value Comparison')
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.5)
plt.tick_params(axis='x', which='minor', bottom=False)
plt.savefig('function-value-comparison-BF.svg')
plt.show()
```
