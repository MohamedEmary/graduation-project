```python
import pandas as pd
import matplotlib.pyplot as plt
import scienceplots
import matplotlib.ticker as ticker
import numpy as np

plt.rcParams.update({'font.size': 14})
plt.style.use(['science', 'no-latex'])
plt.rcParams['font.family'] = 'Times New Roman'

df = pd.read_csv("./filtered_dataset.csv")
```

```python
plt.figure(figsize=(18, 6))
df['sample_num'] = [str(i) for i in range(0, len(df) + 0)]
plt.plot(df['sample_num'], df['total_time_hybrid'],
         label='Hybrid BF Total Time')
plt.plot(df['sample_num'], df['total_time_aes'],
         label='AES Total Time')
plt.legend()
plt.xlabel('Text Sample')
plt.ylabel('Time')
plt.title('Total Time Taken By Algorithm')
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.5)
plt.gca().xaxis.set_major_locator(ticker.MaxNLocator(12))
plt.tick_params(axis='x', which='minor', bottom=False)
plt.savefig('total-time-comparison.svg')
plt.savefig('total-time-comparison.png')
plt.show()
```

```python
plt.figure(figsize=(18, 6))
df['sample_num'] = [str(i) for i in range(0, len(df))]
plt.plot(df['sample_num'], df['hybrid_encode_time'],
         label='Hybrid BF Encode Time')
plt.plot(df['sample_num'], df['aes_encode_time'],
         label='AES Encode Time')
plt.legend()
plt.xlabel('Text Sample')
plt.ylabel('Time')
plt.title('Encode Time Taken By Algorithm')
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.5)
# Adjust the number of ticks to cover 1000 samples
plt.gca().xaxis.set_major_locator(ticker.MaxNLocator(12))
plt.tick_params(axis='x', which='minor', bottom=False)
plt.savefig('total-encode-time-comparison.svg')
plt.savefig('total-encode-time-comparison.png')
plt.show()
```

```python
plt.figure(figsize=(18, 6))
df['sample_num'] = [str(i) for i in range(0, len(df))]
plt.plot(df['sample_num'], df['hybrid_decode_time'],
         label='Hybrid BF Decode Time')
plt.plot(df['sample_num'], df['aes_decode_time'],
         label='AES Decode Time')
plt.legend()
plt.xlabel('Text Sample')
plt.ylabel('Time')
plt.title('Decode Time Taken By Algorithm')
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.5)
# Adjust the number of ticks to cover 1000 samples
plt.gca().xaxis.set_major_locator(ticker.MaxNLocator(12))
plt.tick_params(axis='x', which='minor', bottom=False)
plt.savefig('total-dencode-time-comparison.svg')
plt.savefig('total-dencode-time-comparison.png')
plt.show()
```

```python
# get the sum of values in the hybrid_encode_time and aes_encode_time columns
hybrid_encode_time_sum = df['hybrid_encode_time'].sum()
aes_encode_time_sum = df['aes_encode_time'].sum()

# get the sum of values in the hybrid_decode_time and aes_decode_time columns
hybrid_decode_time_sum = df['hybrid_decode_time'].sum()
aes_decode_time_sum = df['aes_decode_time'].sum()

# get the sum of values in the total_time_hybrid and total_time_aes columns
total_time_hybrid_sum = df['total_time_hybrid'].sum()
total_time_aes_sum = df['total_time_aes'].sum()
```

```python
# create a bar chart for encode time
plt.figure(figsize=(6, 5))
plt.bar(['Hybrid BF Encode Time', 'AES Encode Time'], [
        hybrid_encode_time_sum, aes_encode_time_sum], width=0.3)
plt.title('Encode Time Over 1000 Samples')
# save the figure as an SVG file
plt.savefig('total_encode_time.svg')
plt.savefig('total_encode_time.png')
plt.show()
```

```python
# create a bar chart for decode time
plt.figure(figsize=(6, 5))
plt.bar(['Hybrid BF Decode Time', 'AES Decode Time'], [
        hybrid_decode_time_sum, aes_decode_time_sum], width=0.3)
plt.title('Decode Time Over 1000 Samples')
plt.savefig('total_decode_time.svg')
plt.savefig('total_decode_time.png')
plt.show()
```

```python
# create a bar chart for total time
plt.figure(figsize=(6, 5))
plt.bar(['Hybrid BF', 'AES'], [total_time_hybrid_sum, total_time_aes_sum], width=0.3)
plt.title('Total Time Over 1000 Samples')
plt.savefig('total_time.svg')
plt.savefig('total_time.png')
plt.show()
```
